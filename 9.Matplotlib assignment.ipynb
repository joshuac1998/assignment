Q. We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:
1. fitting it to the periodic function
2. plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
from scipy.optimize import curve_fit
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
max_temp=[39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
min_temp=[21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
```

```python
months=np.arange(12)
days=np.linspace(0,12,365)

def func(days,avg,amp,offset):
    return (avg + amp * np.cos((days+offset) * 2 * np.pi / days.max()))
par1,cov1=curve_fit(func,months,max_temp,[0,0,0])
par2,cov2=curve_fit(func,months,min_temp,[0,0,0])
```

```python
plt.figure(figsize=(8,6))
plt.plot()
plt.plot(months,max_temp,'ro')
plt.plot(days,func(days,*par1),'r-')
plt.plot(months,min_temp,'bo')
plt.plot(days,func(days,*par2),'b-')
plt.xlabel("Month")
plt.ylabel("Temperature($^\circ$c)")
plt.show()
```

Q. Matplotlib:
This assignment is for visualization using matplotlib:
data to use:
url=https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv
titanic = pd.read_csv(url)
Charts to plot:
1. Create a pie chart presenting the male/female proportion
2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
titanic.head()
```

```python
di=titanic["sex"].value_counts()
plt.pie(di,labels=titanic["sex"].unique()[:2],autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(titanic['age'], titanic['fare'],c=titanic["sex"].astype("category").cat.codes)
plt.xlabel("Age")
plt.ylabel("Fare Paid")
plt.show()
```
